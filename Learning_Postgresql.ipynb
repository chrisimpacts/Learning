```python
#Sliding Windows

# options:
    # RWS BETWEEN <start> AND <finish>
        # 5 PRECEDING - number of rows before current row
        # 5 FOLLOWING - number of rows after current row
        # UNBOUNDED PRECEDING - include every row since the beginning of the dataset
        # UNBOUNDED FOLLOWING- include every row until the end of the dataset
        # CURRENT ROW - stop calculation at current row

#This query shows the average points from Harry Kane's previous 40 matches (rolling across seasons)
#Then takes the average of the sliding average for each month

query = """
        WITH sub as 
            (SELECT code_player, 
                web_name, 
                season, 
                round, 
                opp_team_name,
                kickoff_time, 
                total_points_gw,

                --SLIDING:
                AVG(total_points_gw) 
                    OVER (PARTITION BY code_player 
                        ORDER BY kickoff_time 
                            ROWS BETWEEN 
                                40 PRECEDING AND CURRENT ROW
                            ) AS sliding_total
                FROM fpl
                )
        SELECT 
            season,
            EXTRACT(YEAR from kickoff_time) as year,
            EXTRACT(MONTH from kickoff_time) as month,
            AVG(sub.sliding_total)
        FROM sub
        WHERE code_player = 78830 --Harry Kane
        GROUP BY season, year, month -- code_player,total_points_gw,kickoff_time, web_name, season, round, opp_team_name, sub.sliding_total
        ORDER BY season, year, month --code_player, season ASC, round ASC
        ;
        """
with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```

```python
# SQL window functions perform a calculation across a set of table rows that is somehow related to the current row
```

```python
#This query finds low scoring matches among the teams which have a high avg xg in their games
query = """
        SELECT
            spi.team_name,
            spi.opp_team_name,
            spi.xg,
            spi.opp_xg,
            match_xg.avg_match_xg,
            match_xg.team_match_xg_rank
        FROM spi
        LEFT JOIN (
            SELECT team_name,
            avg_match_xg,
            ROW_NUMBER() OVER
                (ORDER BY avgxg.avg_match_xg DESC) as team_match_xg_rank
            FROM (
                SELECT team_name,
                    avg(xg+opp_xg) as avg_match_xg
                FROM spi
                GROUP BY team_name
                ORDER BY avg_match_xg DESC) as avgxg
        ) as match_xg
        ON spi.team_name = match_xg.team_name
        WHERE team_match_xg_rank < 5
        ORDER BY (xg+opp_xg) ASC
        LIMIT 10
        ;
        """

with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```

### LAG & LEAD

```python
#Find previous and following match's xg diff

query = """
        WITH diff AS
            (SELECT
                date, team_name, opp_team_name,
                (xg-opp_xg) AS xg_diff
            FROM spi
            ORDER BY date ASC)
        SELECT
            date, team_name, opp_team_name, 
            (LAG(xg_diff) OVER (PARTITION BY team_name
                                ORDER BY date ASC)) as xg_diff_lag,
            xg_diff,
            (LEAD(xg_diff) OVER (PARTITION BY team_name
                                ORDER BY date ASC)) as xg_diff_lead
        FROM diff
        ORDER BY xg_diff
        LIMIT 10
        ;
        """
with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```

## PAGING

```python
#Split data into n tiles eg. thirds, quarters, deciles etc.
#Aggregate to show average of each third
query = """
        WITH matches AS(
            SELECT
                (score+opp_score) as match_goals
            FROM spi
            ),
        Thirds AS (
            SELECT
                match_goals,
                NTILE(3) OVER (ORDER BY match_goals DESC) AS Third
            FROM matches
        )
        SELECT
            Third,
            AVG(match_goals) AS avg_goals
        FROM Thirds
        GROUP BY Third
        ORDER BY Third ASC
        ;
        """
with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```

```python
#This query finds the sum of points, disagregated by home/away
#ROLLUP(was_home) >> There are extra rows inserted with was_home=None,team_name=None showing the grand totals for those fields
query = """SELECT
                DISTINCT code_player,
                team_name,
                web_name,
                was_home,
                sum(total_points_gw)
            FROM fpl
            WHERE code_player = 17761
            GROUP BY code_player,web_name, ROLLUP(was_home, team_name)
            ;
    """
with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```

```python
#Count distinct teams for each player
#Find totals for each player for each team and grand totals
#Query finds highest scoring players for ONE team, while having scored points for >2 separate teams
query = """WITH count AS (
                SELECT
                    code_player,
                    web_name,
                    team_name,
                    COUNT (team_name) OVER (PARTITION BY code_player) as team_count,
                    SUM(total_points_gw) as points_total
                FROM fpl
                GROUP BY code_player, web_name, CUBE(team_name)
                ORDER BY sum(total_points_gw) DESC)
            SELECT
                code_player,
                web_name,
                COALESCE(team_name,'All teams') AS team_name,
                points_total
            FROM count
            WHERE team_count > 2
                AND team_name != 'All teams'
            LIMIT 10
            ;
    """
with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```

```python
#Find the longest gap between goals! (For players who scored >10 goals in a season)

query = """
        WITH last_goals AS (
            WITH most_recent AS (
                WITH goals AS (
                    SELECT
                        season,
                        code_player,
                        web_name,
                        kickoff_time,
                        goals_scored_gw,
                        CASE WHEN goals_scored_gw > 0
                            THEN 1
                            ELSE 0 END AS goal_scored,
                        SUM(goals_scored_gw) 
                            OVER (PARTITION BY season, code_player ORDER BY kickoff_time)
                            AS goals_total
                    FROM fpl)
                
                SELECT
                    season,
                    code_player,
                    web_name,
                    kickoff_time,
                    goals_scored_gw,
                    goal_scored,
                    goals_total,
                    MIN(kickoff_time) 
                        OVER (PARTITION BY season, code_player, goals_total
                            ORDER BY kickoff_time
                            ) AS most_recent_goal_date
            FROM goals)

            SELECT
                season,
                code_player,
                web_name,
                kickoff_time,
                goals_scored_gw,
                AGE(kickoff_time, most_recent_goal_date) AS time_since_last_goal
            FROM most_recent
            WHERE goals_total > 10
            ORDER BY code_player, time_since_last_goal)
            
        SELECT
            season,
            code_player,
            web_name,
            MAX(time_since_last_goal) as max_time
        FROM last_goals
        GROUP BY season, code_player, web_name
        ORDER BY max_time DESC
        LIMIT 10
        ;
    """
with conn.cursor() as cur:
    dat = pd.read_sql_query(query, conn)
dat
```
