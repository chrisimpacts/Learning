```python
print("Players with the highest xGI, also having high defensive contributions")
view = db_creator.run_sql(f"""
SELECT
                          player_name_id AS p,
                          SUM(minutes) AS sum_mins,
                          SUM(defensive_contribution) AS sum_dc,
                          SUM(expected_goal_involvements) AS sum_xgi
    FROM playergw
    WHERE season = {current_season}
GROUP BY p
HAVING SUM(defensive_contribution) > 150
    AND SUM(minutes) > 2000
ORDER BY sum_xgi DESC
LIMIT 10
;""")
view
```
